```python
import pandas as pd
```

```python
src_data = pd.read_csv(r'/Users/liubo/Desktop/data_project/pokemon.csv')
```

```python
new_data = src_data[['name','generation','attack','defense','capture_rate','percentage_male']]
```

```python
new_data['战斗力'] = (new_data['attack'] + new_data['defense'])/2
```

```python
new_data = new_data.drop(['attack','defense'],axis = 1)
```

```python
new_data
```

```python
new_data = new_data.rename(columns = {'percentage_male':'女生比率'})
```

```python
new_data['女生比率'] = 100 - new_data['女生比率']
```

```python
new_data
```

```python
df = new_data.groupby('generation').agg('mean')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(1)
plt.scatter(df.index,df['女生比率'])
plt.figure(1)
plt.scatter(df.index,df['战斗力'])
plt.show()
```
